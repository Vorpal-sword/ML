```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import fetch_openml

#   MNIST
mnist = fetch_openml('mnist_784', version=1, as_frame=True)
df = mnist.frame  #   DataFrame

#     float
df.iloc[:, :-1] = df.iloc[:, :-1].astype(float)
df.iloc[:, -1] = df.iloc[:, -1].astype(int)  #    int

#    
def plot_mnist_samples(df, num_samples=10):
    plt.figure(figsize=(12, 6))
    
    for i in range(num_samples):
        plt.subplot(2, num_samples // 2, i + 1)
        
        #   
        index = np.random.randint(0, len(df))
        
        #    
        image = df.iloc[index, :-1].to_numpy().reshape(28, 28)
        label = df.iloc[index, -1]  #      ()
        
        # 
        plt.imshow(image, cmap="gray")
        plt.title(f"Label: {label}")
        plt.axis("off")
    
    plt.show()

#     10  
plot_mnist_samples(df, 10)
```

```python
import numpy as np

L = 100
l = 50 
n_experiments = 1000000
def experiment(L, l):
    a = np.random.uniform(0, L)
    b = np.random.uniform(0, L)
    return abs(a - b) < l

results = [experiment(L, l) for _ in range(n_experiments)]

probability = np.mean(results)
probability
```
