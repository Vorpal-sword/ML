```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import fetch_openml

# Завантажуємо датасет MNIST
mnist = fetch_openml('mnist_784', version=1, as_frame=True)
df = mnist.frame  # Перетворюємо в DataFrame

# Перетворюємо всі пікселі в float
df.iloc[:, :-1] = df.iloc[:, :-1].astype(float)
df.iloc[:, -1] = df.iloc[:, -1].astype(int)  # Мітки класів робимо int

# Функція для візуалізації зразків
def plot_mnist_samples(df, num_samples=10):
    plt.figure(figsize=(12, 6))
    
    for i in range(num_samples):
        plt.subplot(2, num_samples // 2, i + 1)
        
        # Вибираємо випадковий індекс
        index = np.random.randint(0, len(df))
        
        # Отримуємо зображення та мітку
        image = df.iloc[index, :-1].to_numpy().reshape(28, 28)
        label = df.iloc[index, -1]  # Остання колонка — це клас (цифра)
        
        # Відображаємо
        plt.imshow(image, cmap="gray")
        plt.title(f"Label: {label}")
        plt.axis("off")
    
    plt.show()

# Викликаємо функцію для перегляду 10 випадкових зображень
plot_mnist_samples(df, 10)
```
